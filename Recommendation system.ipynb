## 2. Collaborative-Filtering base recommendation system

**Consideration** : we want a dataframe constituting :-
1. Users as COlumn : User who have given 100+ rating determing `Experience User`
2. Books as row : Books on which minimum 10 rating is done `Popular Book`

```python
rating_with_name.groupby('User-ID').count()['Book-Rating'] 
##User-ID column ke basis par grouping kar raha hai. Fir har group me 'Book-Rating' column ke values ko count kar raha hai. 
```

```python
x=rating_with_name.groupby('User-ID').count()['Book-Rating'] >100
```

```python
x
```

```python
Exp_user=x[x].index
```

```python
filtered_rating=rating_with_name[rating_with_name['User-ID'].isin(Exp_user)]
##exp user ka data chorke baki sbka data hata do from rating_with_name
##first task complete
```

```python
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>10
y
```

```python
famous_books= y[y].index
```

```python
famous_books
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating
```

```python
final_rating.drop_duplicates() ## droping if any duplicate is preset 
```

```python
pt=final_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True) ## df showing the fraction of the similar items 
```

```python
pt
```

### Using Cosine Similarity For Similarity score of User 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

**Making Function for recommending best 5 books** :

    To do so we have to calculate : 
    
        1. index value of the book 
        2. similarity of the book ka index and similarity score jo sbse zada ho bakio s.

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]
    
    data=[]
    for i in similar_items:
        item=[]
        temp_df=(books[books['Book-Title']==pt.index[i[0]]])
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
        
    return data
```

```python
recommend("7b")
```
